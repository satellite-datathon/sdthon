```python
df = pd.DataFrame(columns=["cc_p20", "cc_p40", "cc_p60", "cc_p80", "cc_default", "date", "x", "y"])

for file in files:
    #r = pd.read_csv('1536-1024-cloud-2018-07-05.csv')
    r = pd.read_csv(file)

    #r = r.iloc[0,1:]
    #r = pd.DataFrame(r).T
    #r
    r = r.drop(columns=['Unnamed: 0'])
    df = df.append(r, ignore_index=True)


df = df.sort_values(by=['x','y','date'], ascending=True)
```

```python
df['x'] = df['x'].astype('str')
```

```python
df['x'] = df['x'].apply(lambda x: x.replace("npy/",""))
```

```python
df.tail()
```

```python
df['cc_p20_perc'] = 100 * df['cc_p20'] / (512*512)
df['cc_p40_perc'] = 100 * df['cc_p40'] / (512*512)
df['cc_p60_perc'] = 100 * df['cc_p60'] / (512*512)
df['cc_p80_perc'] = 100 * df['cc_p80'] / (512*512)
df['cc_default_perc'] = 100 * df['cc_default'] / (512*512)
```

```python
df['cc_p20_perc'] = df['cc_p20_perc'].apply(lambda x: "{0:.1f}".format(x))
df['cc_p40_perc'] = df['cc_p40_perc'].apply(lambda x: "{0:.1f}".format(x))
df['cc_p60_perc'] = df['cc_p60_perc'].apply(lambda x: "{0:.1f}".format(x))
df['cc_p80_perc'] = df['cc_p80_perc'].apply(lambda x: "{0:.1f}".format(x))
df['cc_default_perc'] = df['cc_default_perc'].apply(lambda x: float("{0:.1f}".format(x)))
```
