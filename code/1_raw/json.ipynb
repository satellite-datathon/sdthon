```python
import json as j
import pandas as pd
import glob

from pkg_resources import resource_filename as fpath
import sys
sys.path.append(fpath(__name__, ""))
fp = fpath(__name__, "..\..\data\phase-02\metadata")

final = pd.DataFrame()
filenames = glob.glob(fp + "\*.json")

names = ['cloud-cover', 'date']

for f in filenames:
    df = j.load(open(f))
    cc = df['cloud-cover']
    date = f.split("\\")[5].split(".")[0]
    r = pd.DataFrame([[cc, date]], columns=names)
    #print(f)
    final = final.append(r)
    
print(final)
```

```python
final
final.to_csv("../../data/1_raw/cloud-cover.csv", header=True, index=False)
```
