```python
def get_geodata(x, y):
    geojson_file = glob.glob(f'../../data/geometries/geo-x{x}-y{y}.geojson')[0]
    #print(geojson_file)
    file = open(geojson_files[0])
    data = json.load(file)    
    return(data)
```

```python
i = 0
x = tiles.loc[i, 'tile_x']
y = tiles.loc[i, 'tile_y']
print(f'{i}: {x}_{y}')
data = get_geodata(x, y)
data
data['features'][0]["properties"]=dict(name=f'x{x}-y{y}')
data['features'][0]["id"] = str(i)
data['features'][0]
data
```

```python
#i = 1
for i in range(1, (tiles.shape[0])):
    #print(i)
    x = tiles.loc[i, 'tile_x']
    y = tiles.loc[i, 'tile_y']
    print(f'{i}: {x}_{y}')
    df = get_geodata(x, y)
    #print(df)
    df['features'][0]["properties"]=dict(name=f'x{x}-y{y}')
    df['features'][0]["id"] = str(i)
    data['features'].append(df["features"][0])    
    #data
```

```python
with open('../../data/phase-02.geojson', 'w') as outfile:
    json.dump(data, outfile)
```
