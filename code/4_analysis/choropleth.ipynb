```python
import json
file = open("../../data/phase-02/phase-02.geojson")
counties = json.load(file)

import pandas as pd
df = pd.read_csv("../../data/phase-02/m.csv", dtype={"id": str})
#print(df.head())
#import plotly.graph_objects as go
from plotly import graph_objs as go

fig = go.Figure(go.Choroplethmapbox(geojson=counties, locations=df.id, z=df.harvested,
                                    colorscale="Viridis", zmin=0, zmax=166448,
                                    marker_opacity=0.5, marker_line_width=0))

fig.update_layout(mapbox_style="carto-positron",
                  mapbox_zoom=8, mapbox_center = {"lat": -20.3168, "lon": 148.356}) # 148.00626945796608, -20.216875427393816
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
